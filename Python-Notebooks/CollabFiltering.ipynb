```python
import pandas as pd 
import numpy as np
import re 
import nltk 
nltk.download('stopwords') 
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer 
corpus = []  

data= pd.read_csv('newdata.csv')

print(data.head())
```

```python
ratings = pd.DataFrame(data.groupby('name')['stars_y'].mean())
```

```python
ratings['number_of_ratings'] = data.groupby('name')['stars_y'].count()
ratings.head()
```

```python
food_matrix = data.pivot_table(index='user_id', columns='name', values='stars_y')
food_matrix.head()
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
AFO_user_rating = food_matrix['Pita Jungle']
contact_user_rating = food_matrix['First Watch']
```

```python
AFO_user_rating.head()
```

```python
contact_user_rating.head()
```

```python
similar_to_air_force_one=food_matrix.corrwith(AFO_user_rating)
```

```python
similar_to_contact = food_matrix.corrwith(contact_user_rating)
```

```python
print(data.columns)
```

```python
corr_contact = pd.DataFrame(similar_to_contact, columns=['Correlation'])
corr_contact.dropna(inplace=True)
corr_contact.head()
corr_AFO = pd.DataFrame(similar_to_air_force_one, columns=['correlation'])
corr_AFO.dropna(inplace=True)
print("Resturants Most Similar to Pita Jungle")
corr_AFO.head()
```
